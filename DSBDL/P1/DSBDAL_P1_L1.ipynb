### Droping rows with all values as NaN

```python
pd_excel = pd_excel.dropna()
```

### Adding a column 
#### Create a Gender column with random male and female colum

```python
gender = []
for i in range(67):
    if np.random.rand() > 0.5:
        pd_excel['DM'].drop_duplicates().count()
        gender.append(1)
    else:
        gender.append(0)
        
```

```python
pd_excel["Gender"] = gender
```

```python
pd_excel['DM'].drop_duplicates().count()pd_excel['Gender']
```

```python
pd_excel["Gender"] = pd_excel["Gender"].replace({1:'Male',0:'Female'})
```

```python
pd_excel
```

```python
pd_excel = pd_excel.replace("AB",np.nan)
```

```python
pd_excel
```

# Inputing data through Dictionary

```python
mydict = {'RN':[1,2,3,4,5,6],'SName':['om','sham','ram','jam','kam','dam']}
df1 = pd.DataFrame(mydict)
df1
```

```python
mydict2 = {'RN':[1,2,3,4,5,6],'Marks':[12,13,15,17,18,18]}
df2 = pd.DataFrame(mydict2)
df2
```

```python
df = pd.merge(df1,df2,on='RN')
df
```

```python
pd.unique(df["Marks"])
```

```python
pd.unique(df["Marks"]).shape
```

```python
df.pivot(index='RN',columns='SName',values='Marks')
```

### Back fill and forward fill in pandas

```python
df4.bfill(axis='rows')
```
