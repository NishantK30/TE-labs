```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.loc[:,["pclass","age","sex","fare","class"]].head()
```

#### Distplot

```python
sns.displot(x=titanic["age"],bins=10, kde=True)
```

```python
sns.jointplot(x=titanic["age"],y=titanic['fare'],kind='scatter')
```

```python
sns.jointplot(x=titanic['age'],y=titanic['fare'],kind='kde')
```

### Rugplot

```python
sns.rugplot(titanic['fare'])
```

### Barplot

```python
sns.barplot(x='sex',y='age',data=titanic)
```
